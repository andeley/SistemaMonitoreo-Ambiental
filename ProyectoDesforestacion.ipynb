**Realizamos implementación del algoritmo DFS Busqueda en profundidad**
Primero importamos la biblioteca networkx para la creación del grafo
Luego agregamos los nodos y conectamos nodos con aristas con pesos (cantidad de deforestación)
Se calcula la cantidad de desorestación por causa para cada municipio
Se buscan los municipios más afectados por causa y se muestran

```python
#Busqueda en Profundidad
import networkx as nx

# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos para cada municipio
municipios = df['MUNICIPIO'].unique()
for municipio in municipios:
    G.add_node(municipio)

# Agregar nodos para cada causa de deforestación
causas = df['CAUSA'].unique()
for causa in causas:
    G.add_node(causa)

# Conectar nodos con aristas que representen la cantidad de deforestación
for municipio in municipios:
    for causa in causas:
        # Calcular la cantidad de deforestación causada por 'causa' en 'municipio'
        deforestacion = df[(df['MUNICIPIO'] == municipio) & (df['CAUSA'] == causa)]['AREA_Ha'].sum()
        if deforestacion > 0:
            G.add_edge(causa, municipio, weight=deforestacion)

for causa in causas:
    print(f"\n\n MUNICIPIO AFECTADOS POR {causa}:")
    for municipio in nx.dfs_preorder_nodes(G, source=causa):
        if municipio in municipios:
            print(municipio)

for causa in causas:
    print(f"\n\n MUNICIPIOS MAS AFECTADOS POR {causa}:")
    municipios_afectados = sorted(G.successors(causa), key=lambda municipio: G[causa][municipio]['weight'], reverse=True)
    for municipio in municipios_afectados:
        print(f"{municipio}: {G[causa][municipio]['weight']} hectáreas")
```

Visualizamos el grafo

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```

Calculamos el municipio más afectado y sus causas

```python
municipio_mas_afectado = None
causas_municipio_mas_afectado = []
max_deforestacion = 0

for municipio in G.nodes:
    deforestacion_total = sum(G[causa][municipio]['weight'] for causa in G.predecessors(municipio))
    if deforestacion_total > max_deforestacion:
        max_deforestacion = deforestacion_total
        municipio_mas_afectado = municipio
        causas_municipio_mas_afectado = list(G.predecessors(municipio))


# Creamos el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(causas_municipio_mas_afectado, [G[causa][municipio_mas_afectado]['weight'] for causa in causas_municipio_mas_afectado])
plt.xlabel('Causas')
plt.ylabel('Área deforestada (hectáreas)')
plt.title(f'Causas de deforestación en {municipio_mas_afectado}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
